```python
import RNN
```

```python
infile = "KARPE.txt"
data = open(infile, 'r').read()# should be simple plain text file
chars = list(set(data))
data_size, vocab_size = len(data), len(chars)
epochs = 50
learning_rate = 1e-2
seq_length = 50
batch_size = 32
hidden_nodes = 256
layers = 2
temperature = 0.6
model = RNN.MyModel(vocab_size,hidden_nodes, layers)
trainer = RNN.Trainer(
    learning_rate,
    batch_size,
    seq_length,
    epochs,
    data,
    "alice_output.txt",
    temperature,
    model
)
trainer.train()
RNN.create_plots(trainer, "task2")
```

```python
f = open("Losses.txt", 'a')
f.write("Test set: " + infile + "\n")
f.write(f"Final train loss: {trainer.TRAIN_LOSS[trainer.global_step -1]}\n")
#f.write(f"Final validation loss: {trainer.VALIDATION_LOSS[trainer.global_step - 300 - trainer.global_step % 300]}\n")
f.write(f"Epochs: {epochs}\nlr: {learning_rate}\nsequence length: {seq_length}\nbatch size: {batch_size}")
f.write(f"Hidden nodes: {hidden_nodes}, hidden layers: {layers}")
f.write("\n"*4  + "-"*15 + "\n")
f.close()
print(trainer.VALIDATION_LOSS)
```
